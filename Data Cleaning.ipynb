```python
#lets reduce our columns for the last time by combining the time columns (date, start, stop time)
#but before we do that lets make sure each unique entry in all columns has the same formatting
for col in df:
    print(df[col].describe())
    print(df[col].unique())
    print("\n")
#we can see by inspection that in lecture date column one entry is in a different format from the rest, lets fix this
```

```python
df['Lecture date'].replace('18/10/2019', '18-Oct-19', inplace=True) #correct format
```

```python
#lets make new columns on start and stop time and drop the previous columns
df['Lecture starting time'] = df['Lecture date'] + ' ' +  df['Lecture start time']
df['Lecture end time'] = df['Lecture date'] + ' ' +  df['Lecture stop time']
df.drop(['Lecture date', 'Lecture start time', 'Lecture stop time'], inplace = True, axis = 1)
```

```python
#lets now convert the start and stop columns to a time series datatype
df['Lecture starting time'] = pd.to_datetime(df['Lecture starting time'], format='%d-%b-%y %H:%M')
df['Lecture end time'] = pd.to_datetime(df['Lecture end time'], format='%d-%b-%y %H:%M')
df.loc[1, 'Lecture starting time'].day_name() #confirm that conversion is ok by asking for random weekday
```
